# Data Cleaning

```python
# Drop the 'Cabin' column due to many missing values
df.drop(columns=['Cabin'], inplace=True)
```

```python
# Fill missing values in 'Age' with the median
median_age = df['Age'].median()
df['Age'] = df['Age'].fillna(median_age)
```

```python
# Fill missing values in 'Embarked' with the most common port
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df
```

```python
df.isnull().sum()
```

# Feature Engineering

```python
# Convert 'Sex' into numerical format
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
```

```python
df
```

```python
# Create a new feature 'Title' by extracting titles from names
df['Title'] = df['Name'].apply(lambda x: x.split(',')[1].split('.')[0].strip())
```

```python
# Create a new feature 'FamilySize'
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
```

```python
# One-hot encoding for 'Embarked' and 'Title'
df = pd.get_dummies(df, columns=['Embarked', 'Title'], drop_first=True)
```

```python
df
```

#  Feature Selection and Scaling

```python
# Define the features and target variable
X = df.drop(['PassengerId', 'Name', 'Ticket', 'Survived'], axis=1)
y = df['Survived']
```

```python
# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# Train-Test Split

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

# Model Training

```python
# Logistic Regression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
# Predictions
y_pred_logreg = logreg.predict(X_test)
```

```python
# Evaluation
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_logreg))
print(confusion_matrix(y_test, y_pred_logreg))
print(classification_report(y_test, y_pred_logreg))
```

```python
# Random Forest Classifier
rfc = RandomForestClassifier(n_estimators=100, random_state=42)
rfc.fit(X_train, y_train)
```

```python
# Predictions
y_pred_rfc = rfc.predict(X_test)
```

```python
# Evaluation
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rfc))
print(confusion_matrix(y_test, y_pred_rfc))
print(classification_report(y_test, y_pred_rfc))
```

```python
# Cross-validation for Logistic Regression
cv_logreg = cross_val_score(logreg, X_scaled, y, cv=5)
print("Logistic Regression CV Accuracy:", cv_logreg.mean())
```

```python
# Cross-validation for Random Forest
cv_rfc = cross_val_score(rfc, X_scaled, y, cv=5)
print("Random Forest CV Accuracy:", cv_rfc.mean())
```
